```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X =[]
Y = []

for line in open('D://machine_learning_examples-master//linear_regression_class/data_1d.csv'):
    x,y = line.split(',')
    X.append(float(x))
    Y.append(float(y))
```

```python
# converto numpy aaary
X = np.array(X)
Y = np.array(Y)
```

```python
#plot the data
%matplotlib inline
plt.scatter(X,Y)
plt.show()
```

```python
#apply the equation 
denominator = X.dot(X)-X.mean()*X.mean()
a=(X.dot(Y)-X.mean()*Y.mean())/denominator
b= (Y.mean()*X.dot(X)-X.mean()*X.dot(Y))/denominator
```

```python
print(a)
print(b)
```

```python
# predicted 
Yhat = a*X +b
plt.scatter(X,Y)
plt.plot(X,Yhat)

plt.show()
```

```python
X = []
Y = []

for line in open('D://machine_learning_examples-master//linear_regression_class/data_1d.csv'):
    x,y = line.split(',')
    X.append(float(x))
    Y.append(float(y))
```

```python
X = np.array(X).reshape(-1, 1)
Y = np.array(Y).reshape(-1, 1)
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test =train_test_split(X,Y,test_size=0.3,random_state=43)
```

```python
from sklearn.linear_model import LinearRegression
simple_linear_Regression = LinearRegression()
```

```python
simple_linear_Regression.fit(X_train,y_train)
```

```python
prediction = simple_linear_Regression.predict(X_test)
```

```python
plt.scatter(X_train,y_train,color='red')
plt.plot(X_test,prediction)
plt.show()
```

```python
#from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
#print(confusion_matrix(y_test,prediction))
#print(classification_report(y_test,prediction))
#print(accuracy_score(y_test,prediction))
```

```python
intercept = simple_linear_Regression.intercept_
print(intercept)
```

```python
slope = simple_linear_Regression.coef_
print(slope)
```

```python
simple_linear_Regression.score.
```

```python
from sklearn.metrics import mean_squared_error,r2_score
print(mean_squared_error(y_test,prediction))
```

```python
print(r2_score(y_test,prediction))
```
